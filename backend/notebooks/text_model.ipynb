```python
from googlesearch import search
import newspaper
config = newspaper.Config()
config.browser_user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/91.0.4472.124 Safari/537.36'
def get_google_links(query):
    links=[]
    for url in search(query,2):
        links.append(url)
    return links    
```

```python
query="modi reached pakistan as of 10 july 2025"

response=get_google_links(query)
```

```python
response
```

```python
from newspaper import Article
def extract_article_content(urls):
    content=[]
    for url in urls:
        try:
            article=Article(url,config=config)
            article.download()
            article.parse()
            content.append(article.text)
        except:
            content.append("Could not fetch content")
    return content        
data=extract_article_content(response)
```

```python
print(data)
```

```python
import re 
def preprocess_articles(articles):
    cleaned=[]
    for text in articles:
        text=text.lower()
        text=text.encode('ascii',errors='ignore').decode()
        text=re.sub(r'\s+',' ',text).strip()
        if text:
            cleaned.append(text)
    for data in cleaned:
        if data== 'could not fetch content':
            cleaned.remove(data)       
    return cleaned    
content=preprocess_articles(data)
```

```python
content
```

```python
data='\n\n'.join(content)
data=data.split()
data=' '.join(data[:3400])
data
```

```python
systemprompt="You are a fast news checking ai here is provided the data from websited scraped and after that at last there will be a question from user asking for a news analyze the content and answer only in one word either Real or Fake\n"
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
model=ChatGroq(model='llama3-70b-8192',)
from langchain_core.prompts import ChatPromptTemplate

def check():
    prompt=ChatPromptTemplate.from_messages(
        [
            ('system',systemprompt),
            ('system','{data}\n'),
            ('human','{query}')
        ]
    )
    chain=prompt|model|parser
    response=chain.invoke({'data':data,'query':query})
    return response
```

```python
check()
```
